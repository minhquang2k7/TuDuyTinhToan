```python
def amstrong(x:int) ->bool:
    if x<0:
        return False
    so_chu_so = str(x)
    so_mu = len(so_chu_so)
    tong = 0
    for i in so_chu_so:
        so = int(i)
        tong += so**so_mu
        if tong == x:
            return True
    return False
x = int(input())
print(amstrong(x))

```

```python
def is_magic_num(n:int) -> bool:
    while n > 9:
        tong = 0
        while n > 0:
            tong += n % 10
            n //= 10
        n = tong
    return n == 1
n= int(input())
print(is_magic_num(n))

```

```python
def are_collinear(x1, y1,x2,y2,x3,y3):
    ve_trai = (y2 - y1)*(x3 -x2)
    ve_phai = (y3 -y2)*(x2-x1)
    return ve_phai == ve_trai
```

```python
def is_point_on_segment(p, a, b):
    """Kiểm tra xem điểm p có nằm trên đoạn thẳng ab hay không"""
    cross_product = (p[1] - a[1]) * (b[0] - a[0]) - (p[0] - a[0]) * (b[1] - a[1])
    if abs(cross_product) > 1e-9: # Không thẳng hàng
        return False
    
    dot_product = (p[0] - a[0]) * (b[0] - a[0]) + (p[1] - a[1]) * (b[1] - a[1])
    if dot_product < 0: return False
    
    squared_length_ba = (b[0] - a[0])**2 + (b[1] - a[1])**2
    if dot_product > squared_length_ba: return False
    
    return True

def is_point_in_quadrilateral(p, quad):
    """
    p: tuple (x, y)
    quad: list of 4 tuples [(x1,y1), (x2,y2), (x3,y3), (x4,y4)]
    """
    x, y = p
    inside = False
    n = len(quad)
    
    for i in range(n):
        p1 = quad[i]
        p2 = quad[(i + 1) % n]
        
        # 1. Kiểm tra xem điểm có nằm trên biên không
        if is_point_on_segment(p, p1, p2):
            return True # Nằm trên biên
        
        # 2. Thuật toán Ray Casting đếm giao điểm
        intersect = ((p1[1] > y) != (p2[1] > y)) and \
                    (x < (p2[0] - p1[0]) * (y - p1[1]) / (p2[1] - p1[1] + 1e-9) + p1[0])
        if intersect:
            inside = not inside

    return inside
```

```python
import math
def goc_abc(A:tuple, B:tuple, C:tuple):
   dxab = B[0]-A[0]
   dyab = B[1]-A[1]
   do_dai_ab = (dxab**2+dyab**2)**0.5

   dxac = C[0]-A[0]
   dyac = C[1]-A[1]
   do_dai_ac = (dxac**2+dyac**2)**0.5

   dxbc = B[0]-C[0]
   dybc = B[1]-C[1]
   do_dai_bc = (dxbc**2+dybc**2)**0.5

   cosabc = (do_dai_ab**2 + do_dai_bc**2 - do_dai_ac**2)/(2*do_dai_ab*do_dai_bc)
   goc = math.acos(cosabc)
   return round(goc, 2)



A= tuple(map(int, input().split()))
B= tuple(map(int, input().split()))
C= tuple(map(int, input().split()))
print(goc_abc(A,B, C))


```
