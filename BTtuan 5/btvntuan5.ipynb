W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python

def so_nguyen_to(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
x = int(input())
print(so_nguyen_to(x))    
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 


```python

def so_hoan_hao(n):
    tong = 0
    for i in range(1, n+1):
        if n % i == 0:
            tong += i
            if tong == n:
                return True
    return False

x = int(input())
print(so_hoan_hao(x))        
        
```

Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài.

```python
def anh_xa(tu_1: str, tu_2: str) -> bool:
    if len(tu_1) != len (tu_2):
        return False
    map_a_to_b = {}
    map_b_to_a = {}
    for char_a, char_b in zip(tu_1, tu_2):
        if (char_a in map_a_to_b and map_a_to_b[char_a] != char_b) or \
        (char_b in map_b_to_a and map_b_to_a[char_b] != char_a):
            return False
        map_a_to_b[char_a] = char_b
        map_b_to_a[char_b] = char_a
    return True
tu_1 = input()
tu_2 = input()
print(anh_xa(tu_1, tu_2))


    
    
    
```

Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|. 


```python
def divide_with_remainder(dividend, divisor):
    q = dividend // divisor
    r = dividend % divisor
    if dividend == divisor*q + r and 0 <= abs(r) < abs(divisor):
        return q, r
    return False
x =int(input())
y =int(input())
print(divide_with_remainder(x, y))   

    
```

W5A14 Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 

```python
def compute(a, b, op):
    match op:
        case '+':
            return a + b
        case '-':
            return a - b
        case '*':
            return a*b
        case '/':
            if b == 0:
                return "Không tính đc"
            return a/b
a = int(input())
b = int(input())
op = input()
print(compute(a, b, op))
        
        
```
