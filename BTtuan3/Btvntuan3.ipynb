```python
n = int(input("Nhập vào một số: "))
so_dao_nguoc = 0

while n > 0:
    chu_so_cuoi = n % 10
    so_dao_nguoc = so_dao_nguoc * 10 + chu_so_cuoi
    n = n // 10

print(f"Số sau khi đảo ngược là: {so_dao_nguoc}")

```
