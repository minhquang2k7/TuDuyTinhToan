Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 


```python
a = int(input())
def so_hoan_hao(a):
    if a <= 1:
        return False
    
    tong_uoc = 0
    for i in range (1, a//2+1):
        if a%i == 0:
            tong_uoc += i
    return tong_uoc == a
        
if so_hoan_hao(a):
    print("hoan hao")
else:
    print("Ko tay")
        

    
```

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)
lớn nhất nhưng nhỏ hơn n

```python
def so_nguyen_duong(n):
    S = 0
    k = 0
    while S + ( k + 1 ) < n:
        k += 1
        S += k
    return k, S

n = int(input())

if n <= 1:
    print("Không có số nào thoả mãn")
else:
    k_found , S_found = so_nguyen_duong(n)
    print(k_found)
    print(S_found)

```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
def so_nguyen_duong(A):
    n = 0
    S = 0.0
    while S <= A:
        n += 1
        S += 1/n
    return n
A= int(input())

n_kq=so_nguyen_duong(A)
print(n_kq)

```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay
không

```python
n = int(input("Nhập n: "))

if n <= 0:
    print(f"{n} không phải là dạng 3^k")
else:
    temp = n
    while temp % 3 == 0:
        temp //= 3
    
    if temp == 1:
        print(f"{n} là dạng 3^k")
    else:
        print(f"{n} không phải là dạng 3^k")
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A = int(input("Nhập số nguyên dương A: "))

f1 = 1
f2 = 1

if A < 1:
    print("Không có số Fibonacci nào thỏa mãn.")
elif A == 1:
    print("Số Fibonacci lớn nhất không vượt quá A là: 1")
else:
    fib_lon_nhat = 1
    
    while f2 <= A:
        fib_lon_nhat = f2  
        f_moi = f1 + f2
        f1 = f2
        f2 = f_moi

    print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {fib_lon_nhat}")
```
