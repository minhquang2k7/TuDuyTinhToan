```python
import math

def uoc_nguyen_to_max(n):
    res = -1
    while n % 2 == 0:
        res = 2
        n //= 2
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        while n % i == 0:
            res = i
            n //= i
    if n > 2:
        res = n
    return res

n=int(input())
print(uoc_nguyen_to_max(n))
```

```python
def palindrome(n):
    s = str(n)
    return s == s[::-1]
def reverse_number(n):
    return int(str(n)[::-1])

n =int(input())
steps = 0

while not palindrome(n):
    rev_n = reverse_number(n)
    n = n +rev_n
    steps += 1 

print(steps)
print(n)    
```

```python
def get_collatz_length(x):
    length = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3*x +1
        length += 1
    return length

n=int(input())
max_len = 0
best_x = 0

for x in range(1, n+1):
    current_len = get_collatz_length(x)
    if current_len > max_len:
        max_len = current_len
        best_x = x

print(f"{best_x}{max_len}")


```
