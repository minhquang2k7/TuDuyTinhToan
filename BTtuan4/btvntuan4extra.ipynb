W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b 


```python
while True:
    try:
        a = int(input())
        b = int(input())
        if a>0 and b>0:
            break
        else:
            print("Nhập số nguyên dương")
    except:
        print("Nhập đúng giá trị") 
print("Toàn bộ ước chung là:", end = " ")
for i in range(1, min(a, b)+1):
    if a%i == 0 and b% i == 0:
        print(i, end =" ")             
```

W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
def is_power_of_2(n):
    if n <= 0:
        return False
    while n %2 ==0:
        n //= 2
    return n ==1
n = int(input())
print(is_power_of_2(n))
```

Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 


```python
def so_nguyen_duong(n):
    S = 0
    k = 0
    while S + (k +1)<n:
        k += 1
        S += k
    return k , S
n = int(input())
if n <= 1:
    print("Không có số")
else:
    print(so_nguyen_duong(n))   
```

Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 


```python
max_n = None
min_n = None
nhap_am1 = False
while not nhap_am1:
    n = map(int, input().split())
    for i in n:
        if i == -1:
            nhap_am1 = True
            break
        if max_n is None:
            max_n = i
            min_n = i
        else:
            if i > max_n:
                max_n = i
            elif i < min_n:
                min_n =i
    if not nhap_am1:
        print("Nhập đến khi nhập -1")            
        
        
        
if max_n is not None:
    print(max_n)
    print(min_n)           
     
```

[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 


```python
A = int(input("Nhập số nguyên dương A: "))

f1 = 1
f2 = 1

if A < 1:
    print("Không có số Fibonacci nào thỏa mãn.")
elif A == 1:
    print("Số Fibonacci lớn nhất không vượt quá A là: 1")
else:
    fib_lon_nhat = 1
    
    while f2 <= A:
        fib_lon_nhat = f2  
        f_moi = f1 + f2
        f1 = f2
        f2 = f_moi

    print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {fib_lon_nhat}")
```

Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
thường và thường thành hoa. 


```python
def doi_ky_tu(chuoi):
    result = ""
    for char in chuoi:
        if char.islower():
            result += char.upper()
        elif char.isupper():
            result += char.islower()
        else:
            result += char
    return result
chuoi = input()
print(doi_ky_tu(chuoi))
```

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số 

```python
chuoi = input()
dem_hoa = 0
dem_thuong = 0
dem_so = 0
for char in chuoi:
    if char.isdigit():
        dem_so += 1
    elif char.isupper():
        dem_hoa += 1
    elif char.islower():
        dem_thuong += 1
print(dem_thuong)  
print(dem_hoa)  
print(dem_so)    
```
