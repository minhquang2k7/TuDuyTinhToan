## Bài 8

```python
a, b, c, d , m, n= map(float, input().split())
D = a*d - b*c 
Dx = m*d -b*n 
Dy = a*n - m*c

if D != 0:
    x = Dx / D
    y = Dy / D
    print(f"Hệ có nghiệm duy nhất: {x},{y}")
else:
    if Dx == 0 and Dy == 0:
        print("Hệ có vô số nghiệm")
    elif Dx != 0 or Dy != 0:
        print("Hệ vô nghiệm")
```
